 # While loop

```python
i = 1
n = 5 
while i <= n:
    print(i)
    i = i+1
```

```python
# Table
i = 1
j = 10
n = int(input("Enter a number : "))
while i <= j:
    print(i ,"x",n,"=",i*n)
    i = i+1
```

# For loop

```python
fruits = ["Apple","Banana","Cherry"]
for x in fruits:
     print(x)
```

```python
# Looping through a String
for x in "Banana":
    print(x)
```

```python
for i in range(5):
    print(i)
```

```python
for i in range(5,50,10):
    print(i)
```

```python
for i in range(1,11):
    print(i)
```
