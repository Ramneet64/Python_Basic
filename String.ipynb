# Strings

```python
a = "Hello" # Double inverted commas
print(a)
```

```python
a = 'Hello' # Single inverted commas
print(a)
b = "My name is Ramneet Kaur.I am from punjab India .I am a data analyst as well as I am a Java Developer"
print(b)
```

```python
# Multi line String
a = """My name is Ramneet Kaur
 I am from punjab India 
 I am a data analyst 
 as well as I am a Java Developer """
print(a)
```

# Indexing

```python
a = "Hello Students"
print(a[11])
print(a[-2])
```

# Slicing

```python
s = "Hello Students how are you ?"
print(s[0:5])
```

```python
s = "Hello Students how are you ?"
print(s[6:18])
```

```python
# skip letter
s = "Hello Students how are you ?"
print(s[0:5:2])
```

```python
# Reverse String
s = "Ramneet"
print(s[::-1])
```

```python
# Length of string
s = "Hello Students how are you ?"
print(len(s))
```

```python
# upper method
a = "Hello Students how are you ? "
print(a.upper())
```

```python
# Lower method 
s = "Hello Students how are you ?"
print(s.lower())
```

```python
# Replace
s = "Hello Students how are you ?"
print(s.replace("Students", "Ramneet"))
```

```python
x = "Ramneet Kaur Weds Kirandeep Singh"
print(x)
print(x.replace("Weds","Wife of"))
```

```python
# Find index number
a = "hlo"
b = a.find("l")
print(b)
```
