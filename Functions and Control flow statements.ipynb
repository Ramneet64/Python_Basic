# if Condition

```python
# Write a program to calculate the percentage of a student in math science social english and hindi

x = input("Enter name of Student")
a = int(input(" English Marks : "))
b = int(input(" Hindi Marks : "))
c = int(input(" Math Marks : "))
d = int(input(" Science Marks : "))
e = int(input(" Social Marks : "))
n =((a+b+c+d+e)/500)*100
print(x ,"'s Total Percentage : ",n) 
if n > 90:
    print("Grade : A")
if n >= 75 and n < 90:
    print("Grade : B")
if n >= 65 and n < 75:    
    print("Grade : c")
```

# if - elif - else

```python
""" Accept the city and print its monument
Delhi = Red Fort 
Agra = Taj Mahal
Jaipur = Jalmahal
Punjab = Golden temple
Hyderabad = Charminar
Other = Record not Found"""
```

```python
print("Select City 1. Delhi, Agra, Jaipur, Punjab, Hyderabad, Other ") 
city = input("Enter City")
if city == "Delhi":
    print("Red Fort")
elif city == "Agra":
    print("Taj Mahal")
elif city == "Jaipur":
    print("Jalmahal")
elif city == "Punjab":
    print("Golden Temple")    
elif city == "Hyderabad":
    print("Charminar")    
else :
    print("No Record")    
```
