```python
marks = 87
if marks>=90:
    print("you will get mobile phone")
else:
    print("Study more")
    print("Sutdy hard")
```

# if - elif-else statement

```python
marks = 87

if marks >=90:
    print("you can go to trip")
elif marks >=80 and marks <90:
    print("you will get a new phone")
elif marks >=70 and marks < 80:
    print("you will get a new book")
else :
    print("you will not get your phone back")
```

# nested if

```python
marks = 87
if marks >= 80:
    print("get phone")
    if marks >=95:
        print("go trip")

else:
    print("no phone")
```
