# NUMPY

```python
import numpy as np
```

```python
a = np.array([10,20,30])
print(a)
print(type(a))
```

```python
arr = np.array([[10,20,30,40],[50,60,70,80]])
print(arr)
```

```python
arr = np.array([10,20,30,40])
print(arr)
print(arr[0:3])
```

```python
arr = np.array([[10,20,30,40],[50,60,70,80]])
print(arr[0:2,0:3])
print(arr[0,1:3])
print(np.shape(arr))
print(np.size(arr))
print(np.ndim(arr))
```

```python
a = [[30,40,40],[40,30,10]]
arr = np.array(a)
print(arr)
print(arr.shape)
print(len(arr))
print(np.size(arr))
print(type(arr))
print(arr.dtype)
print(arr.astype(float))
```

```python
arr1 = np.array([[30,40],[20,10]])
arr2 = np.array([[30,20],[20,20]])
print(arr1+arr2)
print(np.add(arr1,arr2))
```

```python
arr1 = np.array([[30,40],[20,10]])
arr2 = np.array([[30,20],[20,20]])
print(arr1-arr2)
print(np.subtract(arr1,arr2))
```

```python
arr1 = np.array([[30,40],[20,10]])
arr2 = np.array([[30,20],[20,20]])
print(arr1*arr2)
print(np.multiply(arr1,arr2))
```

```python
arr1 = np.array([[30,40],[20,10]])
arr2 = np.array([[30,20],[20,20]])
print(arr1/arr2)
print(np.divide(arr1,arr2))
```

```python
arr1 = np.array([3,4,2,1])
arr2 = np.array([3])
print(np.power(arr1,arr2))
```

```python
arr1 = np.array([27,64,8,1])
print(np.sqrt(arr1))
```

```python
arr1 = np.array([30,40,50])
arr2 = np.array([5,5,3])
print(np.concatenate([arr1,arr2]))
```

```python
arr1 = np.array([30,40,50])
print(np.append(arr1,[90]))
print(np.sort(arr1))
print(np.where(arr1 ==40))
```

```python
arr1 = np.array([30,40,50])
print(np.sum(arr1))
print(np.min(arr1))
print(np.max(arr1))
print(np.size(arr1))
print(np.mean(arr1))
```
