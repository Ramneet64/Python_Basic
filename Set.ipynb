```python
# creating a set
a = {"Animal", "Ball" , "Cat"}
print(a)
print(type(a))
```

```python
a = {"Animal", "Ball" , "Cat","Animal", "Ball" , "Cat","Animal", "Ball" , "Cat"}. # no duplicate
print(a)
```

```python
# True and 1 same value
a = {1,2,3,4,4,True}
print(a)
print(len(a))
```

```python
# data type
s1 = {1,2,3,4}
s2 = {1.2 , 2.3, 3.4, 4.5}
s3 = {"a","b","c","d"}
s4 = {True,False}
print(s1,s2,s3,s4)
```

```python
# create tuple in set
a = {(1,2,3,4)}
print(a)
```

```python
# set constructor
x = set(("a","b","c"))
print(x)
print(type(x))
```

```python
# loop
x = {"A","b",1,3.4,"c"}
for i in x:
    print(i)
```

```python
# check item in set 
a = {"Data","Python","Code"}
print("Code" in a)
```

```python
# add item
a = {"a","b","c"}
a.add(100)
print(a)
```

```python
# merge two set
a = {"a","b","c"}
b = {"Data","Python","Code"}
a.update(b)
print(a)
```

```python
# remove
x = {'Data', 'a', 'Python', 'Code', 'b', 'c'}
x.remove("a")
print(x)
```

```python
# join two sets
a = {"a","b","c"}
b = {"Data","Python","Code"}
c = a.union(b)
print(c)
```

```python
# find min and max value
n = {3,4,6,7,2,3,8}
print(n)
print(max(n))
print(min(n))
```
