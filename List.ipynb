```python
# store any type of data
list1 = ["Apple","Banana","Mango"] # String
list2 = [1,2,3,4,5,6] # Int
list3 = [1.2, 3.4 , 5.6] # Float
list4 = [True , False] # Boolean
print(list1)
print(list2)
print(list3)
print(list4)
```

```python
# access item
thislist = [1,2,3,4,5,6,7,8,9,0]
print(thislist[8])
print(thislist[-8])
```

```python
# Slicing
thislist = [1,2,3,4,5,6,7,8,9,0]
print(thislist[0:6])
```

```python
# change items
a = ["Apple", "Banana" , "Mango" , 12,3,4,5,6,7,6.5]
a[1] = "Kela"
a[-1] = 8
print(a)
```

```python
a = ["Apple", "Banana" , "Mango" , 12,3,4,5,6,7,6.5]
a[4:6] = "Kela","Amb"
print(a)
```
