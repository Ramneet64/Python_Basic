```python
# change value
dict = { "ID" : 123,
     "Color" : "Black",
     "Year" : 1997,
     "a" : [1,2,3,4,5],
      "b" : (2.3 ,4.5, 6.7),
        "c" : True
    }
print(dict)
dict["ID"] = 1234
print(dict)
```

```python
# add value
dict = { "ID" : 123,
     "Color" : "Black",
     "Year" : 1997,
     "a" : [1,2,3,4,5],
      "b" : (2.3 ,4.5, 6.7),
        "c" : True
    }
print(dict)
dict.update({"name" : "Ramneet"})
print(dict)
```

```python
# delete
dict = { "ID" : 123,
     "Color" : "Black",
     "Year" : 1997,
     "a" : [1,2,3,4,5],
      "b" : (2.3 ,4.5, 6.7),
        "c" : True
    }
print(dict)
dict.pop("c")
print(dict)
```

```python
# concatenate
d1 = {1:10, 2:20}
d2 = {3:30, 4:40}
d3 = {5:50, 6:60}
d4 = {}
for d in (d1,d2,d3):
    d4.update(d)
print(d4)    
```

```python
# check whether a given key already exsits in a dictionary
d = {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}
def is_key_present(x):
    if x in d:
        print("yes")
    else:
        print("No")
is_key_present(5)
is_key_present(9)
```
